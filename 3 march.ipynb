````python


**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

1. **Histogram**: A histogram is used to visualize the distribution of a single variable. It displays the frequency or probability of different values within a continuous or discrete dataset.

2. **Bar Plot**: Bar plots are useful for comparing the values of different categories or groups. They represent data with rectangular bars, and the height or length of each bar corresponds to the data's value.

3. **Scatter Plot**: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a data point, making it easy to spot patterns or correlations.

4. **Box Plot (Box-and-Whisker Plot)**: Box plots are employed to display the distribution of a dataset and identify outliers. They show the median, quartiles, and potential extreme values, helping to understand data spread and skewness.

5. **Pair Plot**: A pair plot, also known as a scatterplot matrix, is used to visualize pairwise relationships between multiple variables in a dataset. It's helpful for identifying correlations and patterns among variables.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("Line Plot of FMRI Data")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

This code will create a line plot of "signal" over "timepoint" for different events and regions.

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title("Box Plot of Age by Pclass")

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title("Box Plot of Fare by Pclass")

plt.tight_layout()
plt.show()
```

This code will generate two box plots: one for the relationship between "pclass" and "age," and the other for "pclass" and "fare."

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

This code will produce a histogram of diamond prices, with different cuts distinguished by color.

**Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.pairplot(iris_data, hue='species', diag_kind='kde')
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

This code will generate a pair plot of the Iris dataset, where different species are color-coded for easy differentiation.

**Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data for heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d', linewidths=1)
plt.title("Flights Heatmap")
plt.show()
```

This code will produce a heatmap of the "flights" dataset, showing the number of passengers over different months and years.
````
